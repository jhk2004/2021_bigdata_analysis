```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.random.rand(50)  # [0, 1) 0이상 1미만
y = np.random.rand(50)
plt.scatter(x, y)
```

```python
plt.figure(figsize = (10, 10))
plt.scatter(x, y, marker = '*')
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x,y, marker='s')
plt.show()
```

```python
# Color 지정하기
plt.scatter(x, y, c= 'r')
plt.show()
```

```python
# 점마다 Color를 지정하기
x = [1,2,3]
y = [1,2,3]
colors = ['r','g','b']
plt.scatter(x,y,c = colors)
plt.show()
```

```python
# ColorMap(색상지도)
# 데이터를 기반으로 색상 표현하기
plt.colormaps()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x, y,c = x , cmap = 'Blues')  # c는 강조하고자하는 부분
plt.colorbar()
plt.show()
```

```python
# Marker Size 변경하기
size = np.random.rand(50) * 100 # 0.1 * 100 = 10
plt.scatter(x, y, c=x , cmap= 'Blues', s = size)
plt.show()
```

```python
# 투명도 (alpha, 0~1사이의 값)
plt.scatter(x, y, c=x , cmap='Blues', s = size, alpha = 0.5)
plt.show()
```
