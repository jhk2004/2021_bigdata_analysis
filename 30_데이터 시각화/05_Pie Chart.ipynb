# 05_Pie Chart
* 백분율로 표시, 데이터의 비율을 알아내는데 유용함

```python
import matplotlib.pyplot as plt
data = [1,2,3,4]
plt.pie(data)
plt.show()
```

```python
labels = ['치킨','족발', '팅글', '케익']
plt.pie(data, labels = labels)
plt.show()
```

```python
plt.pie(data, labels= labels, autopct='%.1f%%')
```

```python
explode = [0.2,0,0,0]
plt.pie(data, labels = labels, autopct="%.1f%%", explode = explode)
```

```python
colors = ['r', 'g', 'b', 'pink']
plt.pie(data, labels = labels, autopct="%.1f%%", explode = explode, shadow =True, colors = colors)
```
