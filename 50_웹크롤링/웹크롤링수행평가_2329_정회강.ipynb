# 블루투스 키보드 목록 크롤링
* 웹크롤링 대상 사이트 : 네이버 쇼핑 (https://shopping.naver.com)

```python
from bs4 import BeautifulSoup
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

## 데이터 수집 방법
* Selenium을 통한 데이터 수집
1. 네이버 쇼핑에 접속
2. 검색창에 "블루투스 키보드" 검색
3. 옵션 중 "마우스세트" 선택
4. 스크롤을 끝까지 내려서 40개 목록 로딩
5. 페이지 소스 수집

```python
driver_path = 'C:\\JupyterLab\\chromedriver.exe'
driver = webdriver.Chrome(driver_path)
driver.get("https://shopping.naver.com")

element = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH,'//*[@id="autocompleteWrapper"]/input[1]')))
elem = driver.find_element_by_name("query")
elem.clear()
elem.send_keys("블루투스 키보드")
elem.send_keys(Keys.RETURN)

element = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH,'//*[@id="__next"]/div/div[2]/div[2]/div[2]/div/div[4]/div[2]/div/ul/li[5]/a/span')))
mouse_set = "#__next > div > div.style_container__1YjHN > div.style_inner__18zZX > div.filter_finder__1Gtei > div > div:nth-child(5) > div.filter_finder_row__1rXWv > div > ul > li:nth-child(5) > a > span"
driver.find_element_by_css_selector(mouse_set).click()

driver.execute_script("window.scrollTo(0,document.body.scrollHeight)")
time.sleep(3)
html = driver.page_source
driver.close()
```

## 데이터 가공
* BeautifulSoup을 이용하여 제품명, 가격, 찜한 수 를 DataFrame으로 정리


```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(html)
name = []
items = soup.select(".basicList_link__1MaTN")
for item in items:
    name.append(item.get_text())
print(name)
```

```python
prices = soup.select(".price_num__2WUXn")
p =[]
for price in prices:
    p.append(price.get_text())
p
```

```python
likes = soup.select('.basicList_btn_zzim__2MGkM')
l = []
for like in likes:
    l.append(like.get_text()[3:])
l
```

```python
df_data = {
    'name' : name,
    'price': p,
    'like' : l
}

df = pd.DataFrame(data = df_data)
df
```
